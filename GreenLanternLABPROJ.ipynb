```python
def shift_enc(message, key):
    
    encrypt = ""
    
    for char in message:
        ascii_equivalent = (ord(char) + key) % 128
        encrypt += chr(ascii_equivalent)
        
    return encrypt
```

```python
def shift_dec(message, key):
    
    decrypt = ""
    
    for char in message:
        ascii_equivalent = (ord(char) - key) % 128
        decrypt += chr(ascii_equivalent)
        
    return decrypt
```

```python
def vigenere_key(message, keyword):
    
    keys = list(keyword)
    
    if len(message) == len(keyword):
        return keys
    else:
        for i in range(len(message) - len(keyword)):
            additional = keys[i % len(message)]
            keys.append(additional)
        return keys
```
